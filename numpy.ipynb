# 📘 Chapter 3 <b> <u>Indexing & Slicing</u><b>

#### INDEXING = CHOOSE ONE ELEMENTS IN ARRAY
#### SLICING = CHOOSE MULTIPLE ELEMENTS IN ARRAY CHOSE START STOP STEP
#### FANCY INDEXING = CHOOSEMULTIPLE ELEMENTS USING LIST OF INDICES
#### BOOLEN MASKING = GIVE CERTAIN CONDITION AND OPERATE ARRAY 


# <B>INDEXING</B> 
#### ARRAY[INDEX] #1D ARRAY
#### ARRAY[ROW,COLUMN] #2D ARRAY

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr[0])   # First element
print(arr[2])   # Third element
print(arr[-1])  # Last element
```

# <B> SLICING </B>
#### ARRAY[START,STOP, STEP]
#### ARR[START;END] , START TO END -1 
#### NEGATIVE STEP, -1 REVERSE

```python
arr = np.array([10, 20, 30, 40, 50, 60])
print(arr[1:5]) # index 1 to 4
print(arr[:4]) # index 0 to 3
print(arr[::2]) # every secound element
print(arr[::-1]) #reverse value
```

# FANCY INDEXING = CHOOSE MULTIPLE ELEMENTS USING LIST OF INDICES

```python
arr = np.array([10, 20, 30, 40, 50, 60])
print(arr[[0,2,4]])
```

# BOOLEN MASKING = GIVE CERTAIN CONDITION AND OPERATE ARRAY (filtering)

```python
arr = np.array([10, 20, 30, 40, 50, 60])
print(arr[arr > 25])
```

# CONCATE
#### JOIN ARRAYS
#### AXIS 0 ROW WISE VERTICAL STACKING
#### AXIS 1 CLOUMN HORIZONTAL STACKING


```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
new_arr = np.concatenate((arr1, arr2))
print(new_arr)
```

# STACKING

# <B><U> CHAPTER 4 BROADCASTING & VECTORIZATION</B></U>

```python
import numpy as np
prices = np.array([100,200,300,350,400,4555,5555,4567,3333,2222,11122,555])
discount = 10
final_prices = prices-(prices * discount/100)
print(final_prices)
```

### single value brodcasting

```python
arr = np.array([100,200,300])

result = arr * 2
print (result)
```

## BROADCASTING FOR 1D AND 2D ARRAY

```python
matrix = np.array([[1,2,3],[4,5,6]])
vector = np.array([10,20,30])

result = matrix * vector
print(result)
```
